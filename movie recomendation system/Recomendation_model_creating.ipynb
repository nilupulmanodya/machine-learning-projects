```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('datasets/exports/data1.csv')
df_2 = pd.read_csv('datasets/exports/df_2018.csv')
df_3 = pd.read_csv('datasets/exports/df_2019.csv')
df_4 = pd.read_csv('datasets/exports/df_2020.csv')
df_5 = pd.read_csv('datasets/exports/df_2021.csv')
```

```python
df_1.head()
```

```python
df_1 = df_1.rename(columns={'genres':'Genres'})
```

```python
df_1['comb'] = df_1['actor_1_name'] + ' ' + df_1['actor_2_name'] + ' '+ df_1['actor_3_name'] + ' '+ df_1['director_name'] +' ' + df_1['Genres']
```

```python
df = (((df_1.append(df_2,ignore_index=True)).append(df_3,ignore_index=True)).append(df_4,ignore_index=True)).append(df_5,ignore_index=True)
```

```python
df.isna().sum()
```

```python
df = df.fillna('unknown')
```

```python
df.drop_duplicates(subset=['movie_title'],inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Extracting Features

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["comb"])
print("Count Matrix:", count_matrix.toarray())
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
movie_user_likes = "spectre"
def get_index_from_title(title):
    return df[df.movie_title == title].index.values[0]

movie_index = get_index_from_title(movie_user_likes)
```

```python
movie_index
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
similar_movies
```

```python
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["movie_title"].values[0]
i=0
for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0]))
    i=i+1
    if i>15:
        break
```
