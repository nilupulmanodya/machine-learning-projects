assign genres to 2018 movie data sets

```python
def get_genres(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0]['id']
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = ' '
        for i in range (0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        return np.NaN
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    elif " (director/screenplay)" in x :
        return x.split(" (director/screenplay)")[0]
```

```python
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
df_2018
```

```python
def get_actor1(x):
    
    return ((x.split("screenplay);")[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name'] = df_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
df_2018
```

```python
def get_actor_2(x):
    if len(x.split("screenplay); ")[-1].split(", "))<2:
           return np.NaN
    else:
           return (x.split("screenplay);")[-1]).split(", ")[1]
           
```

```python
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor_2(x))
```

```python
df_2018
```

```python
def get_actor_3(x):
    if len(x.split("screenplay); ")[-1].split(", "))<3:
           return np.NaN
    else:
           return (x.split("screenplay);")[-1]).split(", ")[2]
           
```

```python
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor_3(x))
```
