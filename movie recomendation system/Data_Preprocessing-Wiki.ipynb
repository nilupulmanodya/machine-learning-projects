```python
def get_genres(x):
    genres = []
    result = tmdb_movie.search(x)
    if len(result)==0:
        print('movie not found in themoviedb: {}'.format(x))
        return 'unknown'
    
    
    
    else:
        movie_id = result[0]['id']
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = ' '
            for i in range (0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
```

## Implementing function for get 2019/20/21 american movie data from wikipedia and get genres from themoviedb and retutn preprocessed data set

```python
link_2018 = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
link_2019 = 'https://en.wikipedia.org/wiki/List_of_American_films_of_2019'
link_2020 = 'https://en.wikipedia.org/wiki/List_of_American_films_of_2020'
link_2021 = 'https://en.wikipedia.org/wiki/List_of_American_films_of_2021'
```

```python
def movie_data_set(wiki_link):
    df1 = pd.read_html(wiki_link, header=0)[2]
    df2 = pd.read_html(wiki_link, header=0)[3]
    df3 = pd.read_html(wiki_link, header=0)[4]
    df4 = pd.read_html(wiki_link, header=0)[5]
    
    df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
    df['Genres'] = df['Title'].map(lambda x:get_genres(str(x))) #calling get_genres() previously implemented
    df_1 = df[['Title','Cast and crew','Genres']]
    df_1['Genres'] = df_1['Genres'].str.replace(',', '')

    df_1['director_name'] = df_1['Cast and crew'].map(lambda x: get_director(x))#calling get_director()
    df_1['actor_1_name'] = df_1['Cast and crew'].map(lambda x: get_actor1(x))#calling get_actor1()
    df_1['actor_2_name'] = df_1['Cast and crew'].map(lambda x: get_actor_2(x))#calling get_actor_2()
    df_1['actor_3_name'] = df_1['Cast and crew'].map(lambda x: get_actor_3(x)) #calling get_actor_3()
    
    #rename column title to movie_title
    df_1 = df_1.rename(columns={'Title':'movie_title'})
    
    #filtering needed data columns for future model design
    new_df = df_1.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','Genres','movie_title']]
    
    #replacing NaN values
    new_df['actor_2_name'] = new_df['actor_2_name'].replace(np.NaN,'unknown')
    new_df['actor_3_name'] = new_df['actor_3_name'].replace(np.NaN,'unknown')
    
    #change movie titles to lowercase
    new_df['movie_title'] = new_df['movie_title'].str.lower()
    
    #implement combination column
    new_df['comb'] = new_df['actor_1_name']+' '+new_df['actor_2_name']+' '+new_df['actor_3_name']+' '+new_df['director_name']+' '+new_df['Genres']
    
    return new_df
    
    
    
    
    
    
```

#changing function for 2020 and 2021

```python
def movie_data_set_20_21(wiki_link):
    df1 = pd.read_html(wiki_link, header=0)[3]
    df2 = pd.read_html(wiki_link, header=0)[4]
    df3 = pd.read_html(wiki_link, header=0)[5]
    df4 = pd.read_html(wiki_link, header=0)[6]
    
    df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
    df['Genres'] = df['Title'].map(lambda x:get_genres(str(x))) #calling get_genres() previously implemented
    df_1 = df[['Title','Cast and crew','Genres']]
    df_1['director_name'] = df_1['Cast and crew'].map(lambda x: get_director(x))#calling get_director()
    df_1['actor_1_name'] = df_1['Cast and crew'].map(lambda x: get_actor1(x))#calling get_actor1()
    df_1['actor_2_name'] = df_1['Cast and crew'].map(lambda x: get_actor_2(x))#calling get_actor_2()
    df_1['actor_3_name'] = df_1['Cast and crew'].map(lambda x: get_actor_3(x)) #calling get_actor_3()
    
    #rename column title to movie_title
    df_1 = df_1.rename(columns={'Title':'movie_title'})
    
    #filtering needed data columns for future model design
    new_df = df_1.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','Genres','movie_title']]
    
    #replacing NaN values
    new_df['actor_2_name'] = new_df['actor_2_name'].replace(np.NaN,'unknown')
    new_df['actor_3_name'] = new_df['actor_3_name'].replace(np.NaN,'unknown')
    
    #change movie titles to lowercase
    new_df['movie_title'] = new_df['movie_title'].str.lower()
    
    #implement combination column
    new_df['comb'] = new_df['actor_1_name']+' '+new_df['actor_2_name']+' '+new_df['actor_3_name']+' '+new_df['director_name']+' '+new_df['Genres']
    
    return new_df
    
    
    
    
    
    
```
