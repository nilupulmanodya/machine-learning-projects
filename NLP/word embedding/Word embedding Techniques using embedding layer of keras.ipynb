Word embedding Techniques using embedding layer of keras

```python
#need tensorflow 2.0 or above and keras
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences
sent = [
    'This is mango',
    'This is Orange',
    'cup of tea',
    'This is my hometown',
    'old is gold'
]
```

```python
sent
```

```python
#first param for one-hot
voc_size = 1000
```

One hot representation

```python
onehot_rep = [one_hot(words,voc_size) for words in sent]
```

```python
print(onehot_rep)
```

Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embeded_docs = pad_sequences(onehot_rep, padding='pre', maxlen=sent_length)
print(embeded_docs)
```

```python
dim = 15
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embeded_docs))
```
